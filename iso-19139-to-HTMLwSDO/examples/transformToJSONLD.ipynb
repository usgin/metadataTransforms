# Notebook transform example

This notebook gives a simple example tranformating an ISO 19139 data file into JSON-LD 
using this repo's XSL transforms.

```text
usgin/metadataTransforms
"Usgin/Metadatatransforms". Github, 2021, https://github.com/usgin/metadataTransforms. Accessed 27 June 2021.
```

This notebook requires [lxml](https://pypi.org/project/lxml/)

When running in Colab or binder you may need to pull down the XSL file locally or via a URL request. 

````python
import lxml.etree as ET
````

````python
dom = ET.parse("./CDI_ISO19139_full_example_12.2.0.xml")
xslt = ET.parse("../ISO19139ToSDODatasetStandalone1.0.xslt") ## convert to JSON-LD schema.org voc
transform = ET.XSLT(xslt)
newdom = transform(dom)
````
